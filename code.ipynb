Data Collection

```python
import pandas as pd

df = pd.read_csv("data.csv")
# Only save the first 10,000 tweets
df = df.head(10000)
# Save only the tweets in English (i.e. the column for language has the value English)
df = df.loc[df["language"]=="English"]
# Only save the tweets that don't contain a question marked by '?'
df = df[~df["content"].str.contains('\?')]

# Build new column for trump mentions
df["trump_mention"] = df["content"].str.contains(r'\WTrump\W', regex=True)

# Create the required format and save to tsv
dataset = df[['tweet_id', 'publish_date' ,'content', 'trump_mention']].copy()
percentage_mentions_trump = (dataset['trump_mention'].sum() / len(dataset))

dataset['trump_mention'] = dataset['trump_mention'].map({True: 'T', False: 'F'})
dataset.to_csv('dataset.tsv', sep="\t", index=False)
display(dataset)
```

```python
# checking for duplicate tweets, especially coming from the same user
duplicate_values = df.loc[df['content'].duplicated()]
duplicate_values
```

```python
percentage_mentions_trump = round(percentage_mentions_trump, 3)

results = pd.DataFrame([['frac-trump-mentions', percentage_mentions_trump]],columns=['result', 'value'])
results.to_csv("results.tsv", sep="\t", index=False)
```
